```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
marks=[76,54,67,89,34,56,45,67,89]
df=pd.DataFrame({"marks":marks})
print(df)
```

### Mean

```python
mean_value=np.mean(marks)
print(mean_value)
```

### Median

```python
median_value=np.median(marks)
print(median_value)
```

### mode

```python
mode_value=Counter(marks).most_common(2)[0][0]
print(mode_value)
```

### variance

```python
variance_value=np.var(marks)
print(variance_value)
```

### standard deviation

```python
stdandard_dev=np.std(marks)
print(stdandard_dev)
```

### histogram with matplotlib

```python
plt.figure(figsize=(8,5))
plt.hist(marks,color="pink",edgecolor="blue",bins=5)
plt.title("marks")
plt.xlabel("range of marks")
plt.ylabel("No of students")
print(df)
plt.show()
```

### histogram with seaborn

```python
plt.figure(figsize=(8,5))
sns.histplot(df["marks"],bins=5,kde=True,color="green")
plt.title("histogram with seaborn")
plt.xlabel("range of marks")
plt.ylabel("no of std")
plt.show()
```
