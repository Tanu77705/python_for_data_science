```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Bar chart

```python
subjects =["maths", "science", "physics", "chemistry"]
marks=[50,75,45,56]
plt.bar(subjects,marks,color="blue")
plt.title("subjectwise marks")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

fig,axs=plt.subplots(2)
axs[0].plot(x,y1, color="blue")
axs[0].set_title("sine")
axs[1].plot(x,y2, color="red")
axs[1].set_title("cosine")
plt.tight_layout()
plt.show()
```

```python
movies=["3idiots","Dangal","Bahubali","Pushpa","Om shanti Om"]
rating=[4,3,5,4,2]
plt.bar(movies,rating,color="pink")
plt.title("movie rating")
plt.xlabel("movies")
plt.ylabel("rating")
plt.show()
```

```python
routine=["study","sleep","cooking","eating","timepass","work"]
hours=[1,6,4,2,3,2]
plt.pie(hours,labels=routine,autopct="%1.1f%%",startangle=180)
plt.title("zindagi")
plt.axis("equal")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

weather_data={"Day":["Mon","Tue","Wed","Thu","Fri","Sat","Sun"],
              "Temp":[30,32,47,40,39,27,47],
              "humidity":[70,65,70,68,72,65,73]
               }
df=pd.DataFrame(weather_data)
print("weekely weather data:","\n",df)
hottest_day=df[df["Temp"]==df["Temp"].max()]
print("\n","hottest day of the week:","\n",hottest_day)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Day"],df["Temp"],marker="o",color="orange",label="temp")
plt.plot(df["Day"],df["humidity"],marker="s",color="green",label="humidity")

plt.scatter(hottest_day["Day"],hottest_day["Temp"],marker="s",color="blue",label="Hottest day",zorder=10)
plt.title("Weekly weather info")
plt.xlabel("Day of week")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
temps=[28,30,33,45,41,38,36]
max_temp=max(temps)
explode=[0.1 if temp ==max_temp else 0 for temp in temps]
colors=["green" if temp !=max_temp else "red" for temp in temps]

plt.figure(figsize=(6,10))
plt.pie(temps , labels=days,autopct="%1.1f",startangle=180 , explode=explode, colors=colors)
plt.title("Weekly temperature data")
plt.show()
```
